```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
from clean import CleanData
import seaborn as sns
```

```python
data = json.load(open('./data/trending.json', encoding="utf8"))
data = data['collector']
toParse = pd.json_normalize(data)
df = CleanData(toParse)
df.process_the_data()
df.summary_of_data()
```

```python
df_clean = df.dfm
```

```python
# Hypothesis 1 - Create a graph highlighting those with original and those without - Top 50 views first
```

```python
dftop50 = df_clean.head(50)
```

```python
dftop50.shape
```

```python
dftop50['musicMeta.musicOriginal'].value_counts()
```

```python
dftop50_OriginalMusic = dftop50[['id','playCount','shareRate', 'digRate', 'commentRate', 'musicMeta.musicOriginal']]
```

```python
dftop50_OriginalMusic
```

```python
def f(row):
    if row['musicMeta.musicOriginal'] == True:
        val = 'red'
    else: 
        val = 'blue'
    return val
```

```python
# df['C'] = .apply(f, axis=1)
dftop50_OriginalMusic['barColour'] = dftop50_OriginalMusic.apply(f, axis=1)
```

```python
musicdata1 = dftop50_OriginalMusic.sort_values('playCount', ascending=False)
musicdata2 = dftop50_OriginalMusic.sort_values('shareRate', ascending=False)
musicdata3 = dftop50_OriginalMusic.sort_values('digRate', ascending=False)
musicdata4 = dftop50_OriginalMusic.sort_values('commentRate', ascending=False)
```

```python





# create figure
fig = plt.figure(figsize=(9, 6))

# add chart
ax = fig.add_subplot(1, 1, 1)


# plot data 

ax.bar(x='id', height='playCount', color='barColour', data=musicdata1)

# title
ax.set_title('Top 50 View Counts vs Original Music')

# axes
ax.set_ylabel('playCount')

ax.set_yscale('log')

ax.axes.xaxis.set_ticks([])   #hide all the labels

ax.legend(musicdata1['musicMeta.musicOriginal'])

# show plot
plt.show()
```
